# Analizando una Imagen

```python
import urllib.request
urllib.request.urlretrieve('https://raw.githubusercontent.com/Dona505/numpy/main/beagle.jpg', 'beagle.jpg')
import matplotlib.pyplot as plt
img = plt.imread('beagle.jpg')
img.shape

img[10,10]
```

```python
import numpy as np

# Ejercicio 2: Analizar propiedades de la imagen con NumPy
print('Shape de la imagen:', img.shape)
print('Tipo de dato:', img.dtype)
print('Valor maximo de pixel:', np.max(img))
print('Valor minimo de pixel:', np.min(img))
print('Promedio de pixeles:', np.mean(img).round(2))
print('Desviacion estandar:', np.std(img).round(2))
```

```python
# Ejercicio 3: Visualizar la imagen original y sus canales RGB
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

axes[0].imshow(img)
axes[0].set_title('Imagen original')
axes[0].axis('off')

colores = ['Rojos', 'Verdes', 'Azules']

for i in range(3):
    canal = np.zeros_like(img)
    canal[:, :, i] = img[:, :, i]
    axes[i+1].imshow(canal)
    axes[i+1].set_title(f'Canal {colores[i]}')
    axes[i+1].axis('off')

plt.tight_layout()
plt.show()
print('Canales RGB separados correctamente')
```
