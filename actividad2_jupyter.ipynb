## Problema 1: Suma de números pares

Escribe un programa que solicite un número entero positivo `n` y calcule la suma de todos los números pares desde 2 hasta `n`.

```python
def suma_pares(n):
    suma = 0
    for i in range(2, n + 1, 2):
        suma += i
    return suma

numero = 10
resultado = suma_pares(numero)
print(f"La suma de los números pares hasta {numero} es: {resultado}")
```

## Problema 2: Número mayor en una lista

Crea un programa que reciba una lista de números y determine cuál es el valor mayor.

```python
def numero_mayor(lista):
    mayor = lista[0]
    for num in lista:
        if num > mayor:
            mayor = num
    return mayor

datos = [12, 45, 7, 89, 34, 23]
print(f"La lista es: {datos}")
print(f"El número mayor es: {numero_mayor(datos)}")
```

## Problema 3: Verificar palíndromo

Diseña un programa que verifique si una palabra o frase es palíndromo, ignorando espacios y mayúsculas.

```python
def es_palindromo(texto):
    texto_limpio = texto.replace(' ', '').lower()
    return texto_limpio == texto_limpio[::-1]

frase = 'Anita lava la tina'
if es_palindromo(frase):
    print(f"'{frase}' sí es un palíndromo.")
else:
    print(f"'{frase}' no es un palíndromo.")
```
